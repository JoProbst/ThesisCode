```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


result_df = pd.read_csv('../data/chatgpt-27062023-positions-colbert.csv')
# replace position -1 with nan
result_df['position'] = result_df['position'].replace(-1, np.nan)
```

```python
# group by llm and get mean and std of position
llm_df = result_df.groupby('llm').agg({'position': ['mean', 'std']})
```
